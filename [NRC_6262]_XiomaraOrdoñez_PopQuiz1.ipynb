## Preprocesamiento del dataset. Transformaciones previas necesarias para la modelación

```python
# Procesamiento de los datos numericos
"""
    Realizar el llamado a la función para cambiar valores textuales
"""
le = preprocessing.LabelEncoder()
```

```python
#Código que realice la división en entrenamiento y test, de acuerdo con la estretgia de evaluación planeada. Describa cuál es.
"""
    Se realizar la transformación de los valores textuales y numericos
"""
df['Attrition'] = le.fit_transform(df['Attrition'])
df['JobRole'] = le.fit_transform(df['JobRole'])
df['MaritalStatus'] = le.fit_transform(df['MaritalStatus'])
```

```python
"""
    Se transforma las caracteristicas y la recuperación de las mismas
"""
#Transformación de las caracteristicas (mire el apartado Feature engineering del aula virtual)
columnas = ['Attrition','Age','DailyRate','DistanceFromHome','Education','EmployeeCount','RelationshipSatisfaction','StandardHours',
             'TotalWorkingYears','YearsInCurrentRole','YearsAtCompany','JobRole','MaritalStatus']
df_final = df[columnas]
df_final
```

## División del dataset en datos de entrenamiento y datos de test 

```python
#Observar los datos implementando LabelEncoder
df_final
```

```python
#Separar en variables de entrenamiento y test

"""
    Establecer variable objetivo y caracteristicas
"""
columna_x = ['Age','DailyRate','DistanceFromHome','Education','EmployeeCount','RelationshipSatisfaction','StandardHours',
             'TotalWorkingYears','YearsInCurrentRole','YearsAtCompany','JobRole','MaritalStatus']
columna_y = ['Attrition']
X = df_final[columna_x]
y = df_final[columna_y] 
```

```python
#Separar en variables de entrenamiento y Test
"""
    Realizar el correspondiente división de las variables
"""
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
#Observar el tamaño de las variables
"""
    Imprimir las correspondientes tamaños del arreglo
"""
print("Shape X_train: ", X_train.shape)
print("Shape X_test: ", X_test.shape)
print("Shape y_train: ", y_train.shape)
print("Shape y_test: ", y_test.shape)
```

## Modelamiento

```python
#Códo del modelo
"""
    Llamado correspondiente del modelo
"""
svc_lin = SVC(kernel = 'linear', random_state = 0)
```

```python
#Entrenamiento
"""
    Realizar el entrenamiento del modelo
"""
svc_lin.fit(X_train, y_train)
```

```python
#Test
"""
    Predecir el modelo con la variable de test
"""
y_pred= svc_lin.predict(X_test)
```

```python
"""
    Obtención de las metricas correspondientes
"""
print("accurenci:", metrics.accuracy_score(y_test,y_pred))
print("precision:", metrics.precision_score(y_test,y_pred,average='micro'))
print("recall:", metrics.recall_score(y_test,y_pred,average='micro'))
print("F1-score:", metrics.f1_score(y_test,y_pred,average='micro'))
```

## Evaluación del Modelo

Construya un o dos párrafos con los principales hallazgos. Cómo está funcionando el modelo? Calidad en los resultados de predicción. 

```python
print(classification_report(y_test, y_pred, target_names=['Yes', 'No']))
```
